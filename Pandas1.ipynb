## 3 Understanding Series & DataFrame
- **Series**: One-dimensional labeled array, like a column in Excel.
- **DataFrame**: Two-dimensional table, similar to an Excel sheet or SQL table.

### Creating a Series from Python Lists or Dictionaries

````python
import pandas as pd
marks = pd.Series([30,50,20,60,70],
                  index = ['Ronaldo', 'Messi','Virat', 'Dhoni', 'Mbappe'])
print(marks)
````

### Creating a DataFrame

````python
#multiple columns -> dataframe
import pandas as pd

#dataframe are created with python dictionaries
info = pd.DataFrame({
    'name' : ['Rahul', 'Devraj', 'Shreyash', 'Doraemon'],
    'age' : [10,20,30,40],
    'marks' : [0, 0, 90, 0]
})
print(info)

#selecting only the name 
print(info['name'])

#to get specific row we use .loc -> location
print(info.loc[1])

#select any specific value in the row 
print(info.loc[2,'name'])
````

### Creating DataFrame from NumPy Arrays

````python
information = pd.DataFrame([
    ['shreyash', 20, 71],
    ['anushka', 19, 86],
    ['sandy', 21, 76]
], columns=['name', 'age', 'marks'])

print(information)
print(information.loc[2])
print(information.loc[2, 'name'])
````

````python
#Generate a 3X4 matrix with random numbers (1-13) using numpy
#name the columns as [A,B,C,D]

import numpy as np
import pandas as pd

mat = np.random.randint(1,100, size=(3,4))

mat_df = pd.DataFrame(mat, columns=['a', 'b', 'c', 'd'])

print(mat_df)
````
