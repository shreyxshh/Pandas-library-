## 1. Sorting Data
- Use `sort_values(by='col')` to sort by column values.
- Use `ascending=False` for descending order.
- Sort by multiple columns by passing a list.
- `sort_index()` sorts by row or column index.

```python
import pandas as pd

data = {
    'product' : ['pen', 'notebook', 'pencil', 'marker', 'eraser', 'sharpener'],
    'price' : [10, 50, 5, 30, 8 ,12],
    'stock' : [100, 60, 200, 80, 150, 120],
    'rating' : [4.5, 4.7, 4.2, 4.8, 4.3, 4.4]
}
df = pd.DataFrame(data)

#1. sorting by single columns -> price, by default ascending order is followed 
print(df.sort_values(by='price')) #default value for ascending is "true"

#2. sorting by single columns -> rating in descending 
print(df.sort_values(by='rating', ascending=False))

#3. sort my multiple columns -> price and stock
# -> if there are multiple columns wise sorting to be done the most priority will be given to the 
# -> first columns which is passed in the function
# -> multiple columns will be passed in a form of list  
print(df.sort_values(by=['price', 'stock']))

#4. sort index -> in ascneding order 
df.sort_index(ascending=False)
```

## 2. Descriptive Statistics
- `mean()`, `median()`, `mode()` for central tendency.
- `std()` for standard deviation.
- `describe()` for a summary of stats (count, mean, std, min, quartiles, max).

```python
#descriptive stats in only applied on numerical values

#1. mean of numeric columns only
print(df.mean(numeric_only=True)) 

#2. median 
print(df.median(numeric_only=True))

#3. mode
print(df.mode(numeric_only=True))

#4. standard devaition 
print(df.std(numeric_only=True))

#5. describe() -> return complete statistical values of columms
print(df.describe())
```

## 3. Counting Values
- `value_counts()` shows the frequency of unique values in a Series.
- Use `normalize=True` to see percentages instead of counts.

```python
#1. count how many times each objects appears
print(df['product'].value_counts())

#2. shows the percentage instead of counts
print(df['product'].value_counts(normalize=True))

#3. ex : count value of column rating
print(df.rating.value_counts())
```

## 4. Correlation & Covariance
- `corr()` computes correlation between numeric columns.
- `cov()` computes covariance.
- Correlation values range from -1 (negative) to +1 (positive).

```python
#colrrelation is also inly applied in numerical values

#1. correlation btw all numeric column
print(df.corr(numeric_only=True))

#-ve correl = one inc other dec
#+ve correl = one inx other inc

#2. covariance between all numeric columns (not used in this part just to know)
print(df.cov(numeric_only=True))

#3. correlation between two specific columns 
df.price.corr(df.stock)
```
