## 1 Inspecting Data
Quick ways to understand your dataset:
- `head(n)` → first *n* rows (default 5)
- `tail(n)` → last *n* rows
- `info()` → columns, non-null counts, and dtypes
- `shape` → (rows, columns)
- `dtypes` → data type of each column

```python
import pandas as pd

# Create a small DataFrame
a = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Age": [25, 30, 35, 40, 22],
    "Score": [88, 92, 79, 85, 95]
}

b = pd.DataFrame(a)

print(b)

#1. head(input) -> by default 5 rows from the top are return else the desired rows
print(b.head(3))

#2. tail(input) -> by default 5 rows from the buttom are return else the desired rows
print(b.tail(2))

#3. info() -> prints every usefull info about the data 
print(b.info())

#4. shape -> return (rows, cols)
print(b.shape)

#5 dtypes -> returns datatype of the columns
print(b.dtypes)
```

## 2 Selecting Columns & Rows
Ways to access parts of a DataFrame:
- Single column → `df['col']`
- Multiple columns → `df[['col1','col2']]`
- Label-based selection → `df.loc[row_label, col_label]`
- Integer-index selection → `df.iloc[row_idx, col_idx]`
Tip: `:` means "all rows/columns" in that dimension.

```python

#Single columns 
print(b['Age'])

#multiple columns
print(b[["Age","Score"]])

#selecting only row index 2 and col Score
print(b.loc[1, 'Score'])
print(b.loc[3, ['Age', 'Score']])

#DataFrame slicing
print(b.iloc[0:4, 0:2])
```

## 3 Index & Columns Overview
- `df.index` → row labels
- `df.columns` → column names
- `df.values` → underlying NumPy array (read-only; prefer using DataFrame ops)

```python
#1 .index() -> rows kaha se start horhi are kaha end
print(b.index)

#2 .columns() -> returns name of the columns 
print(b.columns)

#3 .values() -> convert the DataFrame into a matrix
print(b.values)
```

## 4 Renaming Columns & Setting an Index
- Rename columns → `df.rename(columns={'old':'new'}, inplace=False)`
- Set a column as index → `df.set_index('column_name', inplace=False)`
- Reset index → `df.reset_index(inplace=False)`
 If you don't use `inplace=True`, remember to assign the result back to a variable.

```python

print(b)

#1 .rename({old_name: new_name}, inplace=True) inplace = True -> changes the og df to the changes which we made
b.rename(columns={'Score' : 'Marks'}, inplace=False)

print(b)

#2 .set_index('col_name', inplace)
c = b.set_index('Name')
print(c)
print(b) # changes will not be applied in og df

#3 .reset_index(inplace)
c.reset_index()
```

## 5 Basic Attributes Recap
- `df.columns` → names of all columns
- `df.index` → row labels
- `df.values` → NumPy array (data only)
These are handy for quick checks and loops (though vectorized ops are preferred).
