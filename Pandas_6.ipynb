# 🐼 Pandas - Class 6: GroupBy & Aggregations
Welcome to **Class 6** of our Pandas series. Today we’ll learn how to group data and perform aggregations.

## 1. Concept of Split–Apply–Combine
- **Split** the data into groups based on some criteria.
- **Apply** a function (e.g., sum, mean) to each group.
- **Combine** the results into a new DataFrame or Series.

This is the core idea behind `groupby`.

```python
import pandas as pd

# New dataset: Employees, their Department, and Salary
data = {
    "Employee": ["John", "Sarah", "Mike", "Anna", "Tom", "Laura", "Steve", "Kate"],
    "Department": ["HR", "IT", "IT", "Finance", "Finance", "HR", "IT", "Finance"],
    "Salary": [50000, 70000, 72000, 65000, 60000, 52000, 75000, 58000]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

## 2. Using `groupby()` with Aggregation Functions
- Use `groupby('col').mean()` or other functions like sum, count, min, max.
- You can also use `agg()` to pass multiple functions at once.

```python
#we use groupby() to group any columns with any other columns using the first as a ref
#groupby(konsa col ko group krna)[iske basis pe].some_agg_func() = both will be the name of th cols  
#to use this groupby() feature we need to add one aggregation function

avg_salary = df.groupby('Department')['Salary'].mean()
avg_salary

#counting the no of members in each department
count_employees = df.groupby('Department')['Employee'].count()
count_employees
```

## 3. Multiple Aggregations on Different Columns
- With `agg()`, you can specify different functions for each column.
- Example: `df.groupby('Dept').agg({'Salary':'mean', 'Age':'max'})`.

```python
import pandas as pd

# Same dataset
data = {
    "Employee": ["John", "Sarah", "Mike", "Anna", "Tom", "Laura", "Steve", "Kate"],
    "Department": ["HR", "IT", "IT", "Finance", "Finance", "HR", "IT", "Finance"],
    "Salary": [50000, 70000, 72000, 65000, 60000, 52000, 75000, 58000],
    "Age": [28, 32, 30, 27, 35, 29, 31, 26]
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

#to get all the aggregation values we use .agg([])
multi_stats = df.groupby('Department')['Salary'].agg(['mean', 'min', 'max'])
multi_stats

#task1 
avg_age = df.groupby('Department')['Age'].mean()
print(avg_age)

#picking two columns to perform aggregation tasks
df.groupby('Department').agg({'Salary' : ['mean', 'min', 'max'],
                              'Age' : ['min', 'max']})
```

## 4. Pivot Tables & Crosstab
- `pivot_table()` summarizes data like Excel pivot tables.
- `crosstab()` shows frequency counts of combinations of factors.
- Both are powerful for summarizing and comparing groups.

```python
#pivot_table(index = main column, values = values which is to be brought, aggfunc = aggregation functions)
print(df.pivot_table(index='Department', values=['Salary', 'Age'], aggfunc='mean')) # not used frequently

#crosstab() :-
df['Age_Group'] = pd.cut(df.Age, bins=[25,30,35,40], labels=['25-30','31-35','35-40'])
#this will create a new column in the df and will be 'Grouped' according to 'Age'
#work for numerical columns only
print(df)
pd.crosstab(df.Department, df.Age_Group)
```

## Mini Practice
1. Create a DataFrame with columns: Department, Employee, Salary, Age, City.
2. Group by Department to get average Salary and max Age.
3. Apply multiple aggregations using agg().
4. Create a pivot_table to see mean Salary by Department and City.
5. Build a crosstab for Department vs City.

```python

# 1. Create a DataFrame with columns: Department, Employee, Salary, Age, City
import pandas as pd
data = {
    "Employee": ["John", "Sarah", "Mike", "Anna", "Tom", "Laura", "Steve", "Kate"],
    "Department": ["HR", "IT", "IT", "Finance", "Finance", "HR", "IT", "Finance"],
    "Salary": [50000, 70000, 72000, 65000, 60000, 52000, 75000, 58000],
    "Age": [28, 32, 30, 27, 35, 29, 31, 26],
    "City" : ['Bhopal','Indore','Pune','Dehli','Bhopal','Bangalore','Dehli','Indore']
}
df = pd.DataFrame(data)
# 2. Group by Department to get average Salary and max Age
# Write your code here
print(df.groupby('Department')['Salary'].mean())
print(df.groupby('Department')['Age'].max())

# 3. Apply multiple aggregations using agg()
# Example: mean Salary, max Age per Department
# Write your code here
print(df.groupby('Department').agg({'Salary' : 'mean', 'Age' : 'max'}))

# 4. Create a pivot_table to see mean Salary by Department and City
# Write your code here
print(df.pivot_table(index='Department', values=['Salary','Age'], aggfunc=['mean', 'count']))##can pass multiple agg function inside a list 

# 5. Build a crosstab for Department vs City
# Write your code here
pd.crosstab(df.Department, df.City)
```
